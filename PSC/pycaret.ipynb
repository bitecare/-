```python
import pandas as pd
from pycaret.classification import *
from time import time
```

```python
train = pd.read_csv("./../DAT/train.csv")
train.head()
test = pd.read_csv("./../DAT/test.csv")
```

```python
train
```

```python
ignore_features=['ID']
categorical_features=['월', '요일','시간', '소관지역','안개', '짙은안개', '번개', '진눈깨비', '서리', '연기/연무', '눈날림', '범죄발생지' ]
numeric_features=['소관경찰서', '사건발생거리', '강수량(mm)', '강설량(mm)', '풍향']
```

```python
cell_start_time = time()
clf = setup(data = train, target='TARGET', ignore_features = ignore_features,
            categorical_features=categorical_features,
            numeric_features=numeric_features, )
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
```

```python
best_models = compare_models(sort='f1', n_select=5, fold=5)
```

```python
total_models = []
models()
```

Gradient Boosting Classifier (gbc)

```python
cell_start_time = time()
model_gbc = create_model('gbc', fold = 5)
total_models.append(model_gbc)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
```

Catboost Classifier (catboost)

```python
cell_start_time = time()
model_catboost = create_model('catboost', fold = 5)
total_models.append(model_catboost)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
```

Light Gradient Boosting Machine (lightgbm)

```python
cell_start_time = time()
model_lightgbm = create_model('lightgbm', fold = 5)
total_models.append(model_lightgbm)
cell_end_time = time()
print("CELL RUN TIME : ",cell_end_time - cell_start_time)
```

Extreme Gradient Boosting (xgboost)

```python
cell_start_time = time()
model_xgboost = create_model("xgboost", fold = 5)
total_models.append(model_xgboost)
cell_end_time = time()
print("CELL RUN TIME : ", cell_end_time - cell_start_time)
```

Ada Boost Classifier(adaboost)

```python
cell_start_time = time()
model_ada = create_model("ada", fold = 5)
total_models.append(model_ada)
cell_end_time = time()
print("CELL RUN TIME : ", cell_end_time - cell_start_time)
```

```python
model_gbc = tune_model(model_gbc, fold=5, optimize = 'F1', choose_better = True)
```
